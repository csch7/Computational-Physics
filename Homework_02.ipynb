### **1. Computing first derivatives:**

```python
# @title
import numpy as np
import matplotlib.pyplot as plt


def sq(x):
  return x**2

def cu(x):
  return x**3

def xp(x):
  return np.exp(-x)

def forward(f, x, h):
  return (f(x+h)-f(x))/h

def central(f, x, h):
  return (f(x+h/2)-f(x-h/2))/h

def five_pt(f, x, h):
  return (-f(x+2*h)+8*f(x+h)-8*f(x-h)+f(x-2*h))/(12*h)

def plot_err(calc, actual, hs, title):
  plt.plot(hs, np.abs(calc-actual), ',')
  plt.yscale('log')
  plt.ylabel('Errors')
  plt.xscale('log')
  plt.xlabel('Step size')
  plt.title(title)
  plt.show()
```

```python
# @title
hs = np.logspace(-20, 5, 5000)

act = 6
fws = forward(sq, 3, hs)
plot_err(fws, act, hs, "Errors for forward difference approximation")
print("Optimal step size: {:.3e}".format(hs[np.argmin(np.abs(fws-act))]))
print("Minimum error: {:.3e}".format(min(np.abs(fws-act))))
print()

cens = central(sq, 3, hs)
plot_err(cens, act, hs, "Errors for central difference approximation")
print("Optimal step size: {:.3e}".format(hs[np.argmin(np.abs(cens-act))]))
print("Minimum error: {:.3e}".format(min(np.abs(cens-act))))
print()


fpts = five_pt(sq, 3, hs)
plot_err(fpts, act, hs, "Errors for five point approximation")
print("Optimal step size: {:.3e}".format(hs[np.argmin(np.abs(fpts-act))]))
print("Minimum error: {:.3e}".format(min(np.abs(fpts-act))))
print()
```

### **2. Derivation of second-order central difference approximation:**

Instead of deriving from the Taylor Approximation, I will just apply the central difference formula twice:
$$\begin{align}f''(x)\approx \frac{f'(x+\frac{h}{2})-f'(x-\frac{h}{2})}{h}\\f'\left(x+\frac{h}{2}\right)\approx \frac{f(x+h)-f(x)}{h}\\f'\left(x-\frac{h}{2}\right)\approx \frac{f(x)-f(x-h)}{h}\\\Rightarrow f''(x)\approx \frac{\frac{f(x+h)-f(x)}{h}-\frac{f(x)-f(x-h)}{h}}{h} = \boxed{\frac{f(x+h)-2f(x)+f(x-h)}{h^2}}\end{align}$$
A case where I would expect it to be exact is for $f(x)=x^2$, where the second derivative is constant (2) for all $x$. Plugging $f(x)=x^2$ into the above approximation, we get that:
$$\begin{align}f(x)=x^2\Rightarrow f''(x) = \frac{(x+h)^2-2x^2+(x-h)^2}{h^2}\\=\frac{x^2+2xh+h^2-2x^2+x^2-2xh+h^2}{h^2}\\=\frac{2x^2+2h^2-2x^2}{h^2}=\frac{2h^2}{h^2} = \boxed{2}\end{align}$$which is the expected solution.

```python
from sys import stderr

def bisection(f, xl, xh, it, tol):
  if xl > xh:
    print("Error: lower bound must be lower than upper bound", file=stderr)
  yl = f(xl)
  yh = f(xh)
  xm, ym = 0, 0
  if (yl > 0 and yh > 0) or (yl < 0 and yh < 0):
    print("Error: Non-single zero found in range, please change low and high bounds", file=stderr)
  for i in range(it):
    xm = (xh+xl)/2
    ym = f(xm)
    if(abs(ym) <= tol):
      break
    print(i, xm, f(xm))
    if min(ym, 0) == 0:
      if(yh < 0):
        yl = ym
        xl = xm
      else:
        yh = ym
        xh = xm
    else:
      if(yh > 0):
        yl = ym
        xl = xm
      else:
        yh = ym
        xh = xm
  return xm, ym

def newton(f, fp, x, it, tol):
  for i in range(it):
    if(abs(x) <= tol): break
    print(i, x, f(x))
    x -= f(x)/fp(x)
  return x, f(x)
```

```python
print(bisection(np.sin, 5, 7, 10, 0))
print(newton(np.sin, np.cos, 7, 10, 0))
```

From this, the Newton-Raphson method clearly converges much more rapidly than the bisection method, reaching a much more optimal solution than bisection ever does in just 3 iterations. A way to reach other zeros is just to start at different points closer to other zeros -- for Newton-Raphson, starting at x=9 instead would converge to a different zero, and starting at [9,10] for bisection would converge to a different zero as well. In order to determine where other roots are, you could sample points randomly on a given interval -- if one point's y-value is positive while another is negative on this interval, there is a root somewhere here -- bisection or Newton-Raphson can then be used to determine where the root is.
